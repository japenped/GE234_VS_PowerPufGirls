```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print("นำเข้าไลบรารีที่จำเป็นเรียบร้อยแล้ว!")
```

```python
# สร้างข้อมูลจำลองสำหรับจำนวนประชากรในสหราชอาณาจักร อิงจากแนวโน้มในลิงก์ Statista
years_uk_2d = np.arange(2000, 2024, 1) # ตั้งแต่ปี 2000 ถึง 2023

# สร้างข้อมูลประชากรจำลองที่มีแนวโน้มเพิ่มขึ้น (ค่าเป็นล้าน)
np.random.seed(42) # เพื่อให้ผลลัพธ์สามารถทำซ้ำได้
base_population = 58.7 # ประชากรเริ่มต้นประมาณปี 2000 (ล้านคน)
population_trend = (years_uk_2d - 2000) * 0.25 # เพิ่มขึ้นประมาณ 0.25 ล้านต่อปี
random_variation = np.random.normal(0, 0.5, len(years_uk_2d)) # เพิ่มความผันผวนเล็กน้อย

population_uk_2d = (base_population + population_trend + random_variation).round(2)

# สร้าง DataFrame
df_uk_population = pd.DataFrame({
    'Year': years_uk_2d,
    'Population_Million': population_uk_2d
})

print("สร้างข้อมูลประชากรจำลองสำหรับสหราชอาณาจักรเรียบร้อยแล้ว!")
print(df_uk_population.head())
```

```python
# สร้างกราฟเส้น 2 มิติ
plt.figure(figsize=(12, 6))
plt.plot(df_uk_population['Year'], df_uk_population['Population_Million'], marker='o', linestyle='-', color='skyblue')

# กำหนดชื่อแกนและชื่อกราฟ
plt.title('UK Total Population Over Time (Simulated Data)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population (Millions)', fontsize=12)

# ปรับแต่งกราฟ
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(years_uk_2d[::2], rotation=45) # แสดงทุก 2 ปี เพื่อไม่ให้แน่นเกินไป
plt.yticks(fontsize=10)
plt.tight_layout()

# แสดงกราฟ
plt.show()
```
